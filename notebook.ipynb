## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Filter for movies only
movies = netflix_df[netflix_df["type"] == "Movie"]

# Filter for movies released in the 1990s
movies_1990s = movies[(movies["release_year"] >= 1990) & (movies["release_year"] <= 1999)]

# Create the duration variable (required for some questions)
duration = int(movies_90s["duration"].mode()[0])

#Count short movies
short_movie_count = int(
    movies_1990s[
        (movies_1990s["genre"] == "Action") &
        (movies_1990s["duration"] < 90)
    ].shape[0]
)

print(f"Most common movie duration in the 1990s: {duration} minutes")
print(f"Number of short Action movies (<90 min): {short_movie_count}")
```

```python
#Distribution of movie durations (1990–1999)

plt.figure(figsize=(10,6))

plt.hist(
    movies_90s["duration"],
    bins=20,
    edgecolor="black",
    alpha=0.8
)

plt.title("Distribution of Movie Durations (1990–1999)", fontsize=14)
plt.xlabel("Duration (Minutes)")
plt.ylabel("Number of Movies")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
#Most common genres among movies released between 1990 and 1999

genre_counts = (
    movies_1990s["genre"]
    .value_counts()
    .rename_axis("Genre")
    .rename("Number of Movies")
    .reset_index()
)

genre_counts.head().style.set_table_styles(
    [{"selector": "table", "props": [("margin-left", "auto"), ("margin-right", "auto")]}]
).hide(axis="index")
```

**Observation**

The most common genre of Movies released between 1990 and 1999 are action, dramas, and comedies, suggesting that character-driven stories and lighter entertainment were especially prominent during these years.

```python
#Top genres among movies released between 1990 and 1999

top_genres = genre_counts.head(10)

plt.figure(figsize=(10,6))

top_genres.plot(
    kind="bar",
    color="steelblue",
    edgecolor="black"
)

plt.title("Top Genres of Movies Released Between 1990 and 1999", fontsize=14)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")

plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

## Conclusion

The movies released between 1990 and 1999 on Netflix usually run just under two hours. 

The average duration is about 115 minutes, while the median duration is 108 minutes, indicating that most films fall close to this range. 

Movie lengths vary widely, from a minimum of 28 minutes to a maximum of 195 minutes, suggesting that while most films cluster around the two-hour mark, there are some much shorter and longer titles in the dataset.

The exploratory analysis of Netflix movies released between 1990 and 1999 shows that the decade includes a substantial number of films with varied genres and runtimes.
